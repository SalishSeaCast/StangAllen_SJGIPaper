```python
atmos   = xr.open_dataset('https://salishsea.eos.ubc.ca/erddap/griddap/ubcSSaSurfaceAtmosphereFieldsV1')
(iY, iX) = places.PLACES['Sand Heads']['wind grid ji'] 

start_date = '2016-01-01'
end_date = '2016-12-31'
start_timestamp = pd.Timestamp(start_date)
end_timestamp = pd.Timestamp(end_date)

udata_2016 = atmos.u_wind.sel(time=slice(start_timestamp, end_timestamp))
u_wind2016 = udata_2016.sel(gridX=atmos.gridX[iX], gridY=atmos.gridY[iY])
vdata_2016 = atmos.v_wind.sel(time=slice(start_timestamp, end_timestamp))
v_wind2016 = vdata_2016.sel(gridX=atmos.gridX[iX], gridY=atmos.gridY[iY])


start_date = '2017-01-01'
end_date = '2017-12-31'
start_timestamp = pd.Timestamp(start_date)
end_timestamp = pd.Timestamp(end_date)

udata_2017 = atmos.u_wind.sel(time=slice(start_timestamp, end_timestamp))
u_wind2017 = udata_2017.sel(gridX=atmos.gridX[iX], gridY=atmos.gridY[iY])
vdata_2017 = atmos.v_wind.sel(time=slice(start_timestamp, end_timestamp))
v_wind2017 = vdata_2017.sel(gridX=atmos.gridX[iX], gridY=atmos.gridY[iY])

start_date = '2018-01-01'
end_date = '2018-12-31'
start_timestamp = pd.Timestamp(start_date)
end_timestamp = pd.Timestamp(end_date)

udata_2018 = atmos.u_wind.sel(time=slice(start_timestamp, end_timestamp))
u_wind2018 = udata_2018.sel(gridX=atmos.gridX[iX], gridY=atmos.gridY[iY])
vdata_2018 = atmos.v_wind.sel(time=slice(start_timestamp, end_timestamp))
v_wind2018 = vdata_2018.sel(gridX=atmos.gridX[iX], gridY=atmos.gridY[iY])

start_date = '2019-01-01'
end_date = '2019-12-31'
start_timestamp = pd.Timestamp(start_date)
end_timestamp = pd.Timestamp(end_date)

udata_2019 = atmos.u_wind.sel(time=slice(start_timestamp, end_timestamp))
u_wind2019 = udata_2019.sel(gridX=atmos.gridX[iX], gridY=atmos.gridY[iY])
vdata_2019 = atmos.v_wind.sel(time=slice(start_timestamp, end_timestamp))
v_wind2019 = vdata_2019.sel(gridX=atmos.gridX[iX], gridY=atmos.gridY[iY])

start_date = '2020-01-01'
end_date = '2020-12-31'
start_timestamp = pd.Timestamp(start_date)
end_timestamp = pd.Timestamp(end_date)

udata_2020 = atmos.u_wind.sel(time=slice(start_timestamp, end_timestamp))
u_wind2020 = udata_2020.sel(gridX=atmos.gridX[iX], gridY=atmos.gridY[iY])
vdata_2020 = atmos.v_wind.sel(time=slice(start_timestamp, end_timestamp))
v_wind2020 = vdata_2020.sel(gridX=atmos.gridX[iX], gridY=atmos.gridY[iY])

start_date = '2021-01-01'
end_date = '2021-12-31'
start_timestamp = pd.Timestamp(start_date)
end_timestamp = pd.Timestamp(end_date)

udata_2021 = atmos.u_wind.sel(time=slice(start_timestamp, end_timestamp))
u_wind2021 = udata_2021.sel(gridX=atmos.gridX[iX], gridY=atmos.gridY[iY])
vdata_2021 = atmos.v_wind.sel(time=slice(start_timestamp, end_timestamp))
v_wind2021 = vdata_2021.sel(gridX=atmos.gridX[iX], gridY=atmos.gridY[iY])

start_date = '2022-01-01'
end_date = '2022-12-31'
start_timestamp = pd.Timestamp(start_date)
end_timestamp = pd.Timestamp(end_date)

udata_2022 = atmos.u_wind.sel(time=slice(start_timestamp, end_timestamp))
u_wind2022 = udata_2022.sel(gridX=atmos.gridX[iX], gridY=atmos.gridY[iY])
vdata_2022 = atmos.v_wind.sel(time=slice(start_timestamp, end_timestamp))
v_wind2022 = vdata_2022.sel(gridX=atmos.gridX[iX], gridY=atmos.gridY[iY])
```

```python
v_wind2016.plot()
```

```python
along_strait2016 = np.cos(np.deg2rad(29))*v_wind2016 + np.sin(np.deg2rad(29))*u_wind2016
along_strait2017 = np.cos(np.deg2rad(29))*v_wind2017 + np.sin(np.deg2rad(29))*u_wind2017
along_strait2018 = np.cos(np.deg2rad(29))*v_wind2018 + np.sin(np.deg2rad(29))*u_wind2018
along_strait2019 = np.cos(np.deg2rad(29))*v_wind2019 + np.sin(np.deg2rad(29))*u_wind2019
along_strait2020 = np.cos(np.deg2rad(29))*v_wind2020 + np.sin(np.deg2rad(29))*u_wind2020
along_strait2021 = np.cos(np.deg2rad(29))*v_wind2021 + np.sin(np.deg2rad(29))*u_wind2021
along_strait2022 = np.cos(np.deg2rad(29))*v_wind2022 + np.sin(np.deg2rad(29))*u_wind2022
```

```python
along_strait = xr.concat([along_strait2016,along_strait2017, along_strait2018,along_strait2019,along_strait2020,along_strait2021,along_strait2022], dim='time')
day_avg_wind = along_strait.resample(time='1D').mean()
day_avg_wind_pd = day_avg_wind.to_dataframe(name='wind')
day_avg_wind_pd = day_avg_wind_pd.drop('gridY',axis=1)
day_avg_wind_pd = day_avg_wind_pd.drop('gridX',axis=1)
```

```python
day_avg_wind_pd.plot();
```

```python
day_avg_wind_pd.to_csv('day_avg_wind_16-22.csv')
```
